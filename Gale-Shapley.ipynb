page 28 og Algorithm Design

initially all m ∊ M and w ∊ W are free
While there is a man m who is free and hasn't proposed to every woman
Choose such a man m
Let w be the highest-ranked woman in m's preference list to whom m has not yet proposed
If w is free then
   (m, w) become engaged
Else w is currently engaged to m′
   If w prefers m′ to m then
      m remains free
   Else w prefers m to m′
      (m,w) become engaged
      m′ becomes free
   Endif
Endif
Endwhile
Return the set S of engaged pairs

```python
"""Populate testdata"""

import random
from typing import Any


class data:
    def __init__(self, ID: int) -> None:
        self.preferences: list[int] = []
        self.dataEntryCount: int = 0
        self.ID = ID

    def populate_preferences(self):
        for random_number in random.sample(
            range(1, self.dataEntryCount + 1), self.dataEntryCount
        ):
            self.preferences.append(random_number)


class dataset:
    def __init__(self) -> None:
        self.entry: dict[int, list[Any]] = {}


dataEntryCount = 10
list_a = dataset()
for id in range(dataEntryCount):
    dataEntry = data(id + 1)
    dataEntry.dataEntryCount = dataEntryCount
    dataEntry.populate_preferences()
    list_a.entry[dataEntry.ID] = [0, dataEntry.preferences]


list_b = dataset()
for id in range(dataEntryCount):
    dataEntry = data(id + 1)
    dataEntry.dataEntryCount = dataEntryCount
    dataEntry.populate_preferences()
    list_b.entry[dataEntry.ID] = [0, dataEntry.preferences]


""" Run Gale-shapley algorithm on created testset"""

matched = 0
while matched != 10:
    matched = 0
    for m in list_a.entry:
        if list_a.entry[m][0] == 0:
            entry_a_preferences = list_a.entry[m][1][0]
            entry_b_current_match = list_b.entry[entry_a_preferences][0]
            w_preferences = list_b.entry[entry_a_preferences][1]

            if entry_b_current_match == 0:
                list_a.entry[m][0] = entry_a_preferences
                list_b.entry[entry_a_preferences][0] = m

            # not matched
            elif entry_b_current_match != 0:
                # check man against womans preferences
                for preference in w_preferences:
                    if preference == entry_b_current_match:
                        break

                    elif preference == m:
                        list_a.entry[entry_b_current_match][0] = 0
                        list_a.entry[m][0] = entry_a_preferences
                        list_b.entry[entry_a_preferences][0] = m
                        break
                    else:
                        continue
            list_a.entry[m][1].pop(0)
    for m in list_a.entry:
        if list_a.entry[m][0] != 0:
            matched += 1

for k in list_a.entry:
    print(f"{k}: list a {list_a.entry[k][0]}, list b {list_b.entry[k][0]}")
```
